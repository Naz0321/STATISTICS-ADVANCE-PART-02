**Q21.What are the different types of ANOVA¬†tests?**

Purpose: Extension of ANOVA when there are multiple dependent variables.

Example: Studying how different diets affect both weight and cholesterol levels simultaneously.

Purpose: Similar to ANOVA but includes one or more covariates to control for confounding variables.

Repeated Measures	1+	Yes	No	No
Mixed ANOVA	1+	Mixed	No.



**Q22.What is the F-test, and how does it relate to hypothesis¬†testing?**


ANS.The F-test is a statistical test that uses the F-distribution to compare variances, or to assess models in the context of hypothesis testing, especially in ANOVA (Analysis of Variance) and regression analysis.


---

Purpose of the F-test:

1. Compare Two Population Variances
Used to test if two populations have equal variances:



H_0: \sigma_1^2 = \sigma_2^2 \quad \text{vs.} \quad H_1: \sigma_1^2 \ne \sigma_2^2

F = \frac{s_1^2}{s_2^2}

2. ANOVA (Analysis of Variance)
Used to test if three or more group means are equal. The F-statistic here is:



F = \frac{\text{variance between groups}}{\text{variance within groups}}

3. Regression Analysis
Tests the overall significance of a multiple regression model:



H_0: \beta_1 = \beta_2 = \dots = \beta_k = 0

H_1: \text{At least one } \beta_i \ne 0


---

How It Relates to Hypothesis Testing:

Null Hypothesis () typically states:

"Variances are equal" (in variance tests),

"All group means are equal" (in ANOVA),

or "None of the predictors are useful" (in regression).


Alternative Hypothesis () asserts that:

Variances differ,

At least one group mean differs,

At least one predictor is significant.


You calculate the F-statistic and compare it to the critical value from the F-distribution, or use the p-value.



---

Key Assumptions:

The data are normally distributed.

Observations are independent.

In ANOVA: Groups have equal variances (homoscedasticity).



---

Example (ANOVA F-test):

Suppose you're testing whether three diets affect weight loss differently:

: All diet groups have the same mean weight loss.

: At least one group mean differs.


If the F-statistic is large and the p-value is small (e.g., < 0.05), you reject  ‚Äî meaning the diets don‚Äôt have equal effects.

```python
# Q5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis¬†testing?

# ANS.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu0=0, mu1=1, sigma=1, alpha=0.05, n=30):
    """
    Visualize Type I and Type II errors for hypothesis testing (Z-test).

    H0: Œº = mu0
    H1: Œº = mu1
    """
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical z for two-tailed test
    z_critical = norm.ppf(1 - alpha/2)

    # Critical values in terms of sample mean
    lower_crit = mu0 - z_critical * se
    upper_crit = mu0 + z_critical * se

    # x-axis range
    x = np.linspace(mu0 - 4*sigma, mu1 + 4*sigma, 1000)

    # Distributions under H0 and H1
    y0 = norm.pdf(x, mu0, se)  # H0 distribution
    y1 = norm.pdf(x, mu1, se)  # H1 distribution

    # Compute Type I error = alpha (already given)

    # Compute Type II error = probability of failing to reject H0 under H1
    beta = norm.cdf(upper_crit, mu1, se) - norm.cdf(lower_crit, mu1, se)

    # Plot distributions
    plt.figure(figsize=(10,6))
    plt.plot(x, y0, label='H0 Distribution (Œº={})'.format(mu0), color='blue')
    plt.plot(x, y1, label='H1 Distribution (Œº={})'.format(mu1), color='green')

    # Shade Type I error regions (alpha/2 each side)
    plt.fill_between(x, 0, y0, where=(x < lower_crit) | (x > upper_crit),
                     color='red', alpha=0.3, label='Type I Error (Œ±)')

    # Shade Type II error region
    plt.fill_between(x, 0, y1, where=(x >= lower_crit) & (x <= upper_crit),
                     color='orange', alpha=0.3, label='Type II Error (Œ≤)')

    # Critical lines
    plt.axvline(lower_crit, color='black', linestyle='--', label='Critical Values')
    plt.axvline(upper_crit, color='black', linestyle='--')

    plt.title('Type I (Œ±) and Type II (Œ≤) Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()

    print(f"Critical values: {lower_crit:.3f}, {upper_crit:.3f}")
    print(f"Type I error (Œ±): {alpha}")
    print(f"Type II error (Œ≤): {beta:.4f}")
    print(f"Power of the test (1-Œ≤): {1-beta:.4f}")

# Example usage
visualize_errors(mu0=0, mu1=1, sigma=1, alpha=0.05, n=30)
```

```python
# Q9.Write a Python function to calculate the confidence interval for a sample mean and explain its¬†significance?

# ANS.

import scipy.stats as st
import numpy as np

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    data (list or array): Sample data
    confidence (float): Confidence level (default = 0.95)

    Returns:
    tuple: (lower_bound, upper_bound)
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = st.sem(data)  # Standard error of the mean

    # t critical value for given confidence and degrees of freedom
    t_critical = st.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_critical * std_err
    return (mean - margin_of_error, mean + margin_of_error)

# Example usage:
sample_data = [12, 15, 14, 10, 13, 16, 14]
ci = confidence_interval(sample_data, confidence=0.95)
print(f"95% Confidence Interval:¬†{ci}")
```

```python
# Q10. Write a Python program to calculate the margin of error for a given confidence level using¬†sample¬†data?

# ANS.

import numpy as np
import scipy.stats as st

def margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for a given confidence level.

    Parameters:
        data (list or array): Sample data
        confidence (float): Confidence level (default=0.95 for 95%)

    Returns:
        float: Margin of error
    """
    n = len(data)
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # sample standard deviation
    alpha = 1 - confidence

    # t-critical value for two-tailed test
    t_critical = st.t.ppf(1 - alpha/2, df=n-1)

    # Margin of error
    moe = t_critical * (std_dev / np.sqrt(n))
    return moe

# Example usage
sample_data = [12, 15, 14, 10, 13, 14, 16, 15, 14, 13]
confidence_level = 0.95

moe = margin_of_error(sample_data, confidence_level)
print(f"Margin of Error at {confidence_level*100}% confidence level:¬†{moe:.4f}")
```

```python
# Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain¬†the¬†process?

# ANS.

def bayesian_inference(prior, likelihood, false_positive_rate):
    # Prior probabilities
    P_H = prior
    P_not_H = 1 - P_H

    # Likelihoods
    P_D_given_H = likelihood
    P_D_given_not_H = false_positive_rate

    # Evidence
    P_D = (P_D_given_H * P_H) + (P_D_given_not_H * P_not_H)

    # Posterior
    P_H_given_D = (P_D_given_H * P_H) / P_D
    return P_H_given_D

# Example values
prior = 0.01       # P(H)
likelihood = 0.99  # P(D|H)
false_positive = 0.05  # P(D|not H)

posterior = bayesian_inference(prior, likelihood, false_positive)
print(f"Posterior probability that the patient has the disease: {posterior:.4f}")
```

```python
# Q16. Implement an F-test using Python to compare the variances of two¬†random¬†samples?

# ANS.

import numpy as np
from scipy.stats import f

# Generate two random samples
np.random.seed(42)  # for reproducibility
sample1 = np.random.normal(loc=50, scale=10, size=30)  # mean=50, std=10
sample2 = np.random.normal(loc=52, scale=15, size=35)  # mean=52, std=15

# Calculate sample variances
var1 = np.var(sample1, ddof=1)  # sample variance (ddof=1 for unbiased)
var2 = np.var(sample2, ddof=1)

# Compute F statistic
if var1 > var2:
    F = var1 / var2
    dfn = len(sample1) - 1  # degrees of freedom numerator
    dfd = len(sample2) - 1  # degrees of freedom denominator
else:
    F = var2 / var1
    dfn = len(sample2) - 1
    dfd = len(sample1) - 1

# Compute p-value (two-tailed)
p_value = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))

# Print results
print("Sample 1 Variance:", var1)
print("Sample 2 Variance:", var2)
print("F-statistic:", F)
print("Degrees of Freedom: ({}, {})".format(dfn, dfd))
print("p-value:", p_value)

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference¬†in¬†variances.")
```

```python
# Q19. Write a Python function to check the assumptions (normality, independence, and equal variance)¬†for¬†ANOVA?

# ANS.

import numpy as np
import pandas as pd
from scipy.stats import shapiro, levene
from statsmodels.stats.stattools import durbin_watson
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(data, group_col, value_col):
    """
    Check ANOVA assumptions: normality, independence, equal variances.

    Parameters:
    data : pandas DataFrame
        Dataset containing the groups and values.
    group_col : str
        Column name for the grouping variable.
    value_col : str
        Column name for the numeric values.

    Returns:
    None (prints results and plots)
    """
    # Fit ANOVA model to get residuals
    model = sm.OLS.from_formula(f"{value_col} ~ C({group_col})", data=data).fit()
    residuals = model.resid

    print("\n--- Checking Normality (Shapiro-Wilk Test) ---")
    stat, p = shapiro(residuals)
    print(f"Shapiro-Wilk Test: W={stat:.4f}, p-value={p:.4f}")
    if p > 0.05:
        print("Residuals seem normally distributed (p > 0.05).")
    else:
        print("Residuals may not be normally distributed (p ‚â§ 0.05).")

    # QQ Plot
    sm.qqplot(residuals, line='s')
    plt.title("QQ Plot of Residuals")
    plt.show()

    print("\n--- Checking Independence (Durbin-Watson) ---")
    dw = durbin_watson(residuals)
    print(f"Durbin-Watson statistic: {dw:.4f} (ideal ~2)")

    print("\n--- Checking Homogeneity of Variance (Levene's Test) ---")
    groups = [group[value_col].values for name, group in data.groupby(group_col)]
    stat, p = levene(*groups)
    print(f"Levene's Test: W={stat:.4f}, p-value={p:.4f}")
    if p > 0.05:
        print("Equal variances assumption seems valid (p > 0.05).")
    else:
        print("Variances may not be equal (p ‚â§ 0.05).")

    print("\n--- Visualization: Residuals Histogram ---")
    sns.histplot(residuals, kde=True)
    plt.title("Histogram of Residuals")
    plt.show()
```

```python
# Q21. Write a Python program to visualize the F-distribution and discuss its use in hypothesis¬†testing?

# ANS.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Parameters for F-distribution
dfn = 5  # degrees of freedom numerator
dfd = 10 # degrees of freedom denominator

# Generate x values
x = np.linspace(0, 5, 500)

# Compute F-distribution PDF
pdf = f.pdf(x, dfn, dfd)

# Plot F-distribution
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, 'b-', label=f'F-distribution (dfn={dfn}, dfd={dfd})')
plt.title('F-distribution PDF')
plt.xlabel('F value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(alpha=0.3)

# Highlight critical region (Œ± = 0.05, right tail)
alpha = 0.05
critical_value = f.ppf(1 - alpha, dfn, dfd)
plt.axvline(critical_value, color='r', linestyle='--', label=f'Critical Value = {critical_value:.2f}')
plt.fill_between(x, 0, pdf, where=(x > critical_value), color='red', alpha=0.3, label='Rejection Region')

plt.legend()
plt.show()
```

```python
# Q22. Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare¬†group¬†means?

# ANS.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import f_oneway
import seaborn as sns

# -----------------------------
# Simulate sample data for 3 groups
# -----------------------------
np.random.seed(42)
group_A = np.random.normal(loc=50, scale=5, size=30)
group_B = np.random.normal(loc=52, scale=5, size=30)
group_C = np.random.normal(loc=55, scale=5, size=30)

# Combine into DataFrame
data = pd.DataFrame({
    'Score': np.concatenate([group_A, group_B, group_C]),
    'Group': ['A']*30 + ['B']*30 + ['C']*30
})

# -----------------------------
# Perform One-Way ANOVA
# -----------------------------
f_stat, p_value = f_oneway(group_A, group_B, group_C)
print(f"F-statistic: {f_stat:.4f}, p-value: {p_value:.4f}")

# -----------------------------
# Visualization: Boxplot of groups
# -----------------------------
plt.figure(figsize=(8, 5))
sns.boxplot(x='Group', y='Score', data=data, palette='Set2')
sns.stripplot(x='Group', y='Score', data=data, color='black', alpha=0.5)

plt.title('Group Comparison using One-Way ANOVA')
plt.xlabel('Group')
plt.ylabel('Scores')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Q23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate¬†the¬†means?

# ANS.

import numpy as np
from scipy.stats import ttest_1samp, ttest_ind
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# Simulate random data from normal distribution
# -----------------------------
np.random.seed(42)

# Population parameters
pop_mean = 50
pop_std = 5

# Sample data
sample1 = np.random.normal(loc=pop_mean, scale=pop_std, size=50)
sample2 = np.random.normal(loc=52, scale=5, size=50)  # slightly different mean

# -----------------------------
# Hypothesis Testing
# 1. One-sample t-test: Is sample1 mean = 50?
# -----------------------------
t_stat1, p_value1 = ttest_1samp(sample1, pop_mean)

print(f"One-Sample t-test:")
print(f"Sample Mean = {np.mean(sample1):.2f}")
print(f"t-statistic = {t_stat1:.4f}, p-value = {p_value1:.4f}\n")

# -----------------------------
# 2. Two-sample t-test: Are means of sample1 and sample2 equal?
# -----------------------------
t_stat2, p_value2 = ttest_ind(sample1, sample2)

print(f"Two-Sample t-test:")
print(f"Sample1 Mean = {np.mean(sample1):.2f}, Sample2 Mean = {np.mean(sample2):.2f}")
print(f"t-statistic = {t_stat2:.4f}, p-value = {p_value2:.4f}")

# -----------------------------
# Visualization
# -----------------------------
plt.figure(figsize=(8,5))
sns.histplot(sample1, color='blue', kde=True, label='Sample 1')
sns.histplot(sample2, color='orange', kde=True, label='Sample 2')
plt.axvline(np.mean(sample1), color='blue', linestyle='--', label=f'Mean1: {np.mean(sample1):.2f}')
plt.axvline(np.mean(sample2), color='orange', linestyle='--', label=f'Mean2: {np.mean(sample2):.2f}')
plt.title('Simulated Normal Distributions for Hypothesis Testing')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Q26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize¬†the¬†results?

# ANS.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Sample data
data1 = [10, 12, 9, 11, 14, 13, 10]
data2 = [8, 9, 6, 7, 10, 8, 9]

alpha = 0.05

# Sample sizes and variances
n1, n2 = len(data1), len(data2)
s1_sq, s2_sq = np.var(data1, ddof=1), np.var(data2, ddof=1)

# F-statistic
F = s1_sq / s2_sq

# Degrees of freedom
df1, df2 = n1 - 1, n2 - 1

# Critical values
F_crit_low = stats.f.ppf(alpha/2, df1, df2)
F_crit_high = stats.f.ppf(1 - alpha/2, df1, df2)

# p-value (two-tailed)
p_value = 2 * min(stats.f.cdf(F, df1, df2), 1 - stats.f.cdf(F, df1, df2))

print(f"Sample Variances: s1^2={s1_sq:.4f}, s2^2={s2_sq:.4f}")
print(f"F-statistic = {F:.4f}")
print(f"Critical region: F<{F_crit_low:.4f} or F>{F_crit_high:.4f}")
print(f"p-value = {p_value:.4f}")
print("Reject H0?", p_value < alpha)

# Visualization of F-distribution
x = np.linspace(0, 5, 500)
y = stats.f.pdf(x, df1, df2)

plt.figure(figsize=(8,5))
plt.plot(x, y, label="F-distribution")
plt.axvline(F, color='red', linestyle='--', label=f"F = {F:.2f}")
plt.axvline(F_crit_low, color='green', linestyle='--', label=f"Lower Crit = {F_crit_low:.2f}")
plt.axvline(F_crit_high, color='green', linestyle='--', label=f"Upper Crit = {F_crit_high:.2f}")
plt.fill_between(x, 0, y, where=(x < F_crit_low) | (x > F_crit_high), color='yellow', alpha=0.3, label='Rejection Region')
plt.title("F-Test for Comparing Two Variances")
plt.xlabel("F value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
# Q27. Perform a Chi-square test for goodness of fit with simulated data and analyze¬†the¬†results?

# ANS.

import numpy as np
from scipy.stats import chisquare

# Simulated observed frequencies (e.g., 4 categories)
observed = np.array([25, 30, 20, 25])

# Expected frequencies (equal distribution assumption)
expected = np.array([25, 25, 25, 25])

print("Observed:", observed)
print("Expected:", expected)

# Perform Chi-square test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square Statistic: {chi_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject H‚ÇÄ: The data does NOT fit the expected distribution.")
else:
    print("Fail to reject H‚ÇÄ: The data fits the expected distribution.")

import matplotlib.pyplot as plt

categories = ['Cat A', 'Cat B', 'Cat C', 'Cat D']

x = np.arange(len(categories))
width = 0.35

plt.bar(x - width/2, observed, width, label='Observed')
plt.bar(x + width/2, expected, width, label='Expected')

plt.xlabel('Categories')
plt.ylabel('Frequency')
plt.title('Observed vs Expected Frequencies')
plt.xticks(x, categories)
plt.legend()
plt.show()
```
